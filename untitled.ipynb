```python
#データ処理ライブラリ
import numpy as np
import pandas as pd

#可視化ライブラリ
import matplotlib.pyplot as plt

#画像処理ライブラリ
import cv2

#keras
import tensorflow as tf
import keras as ks
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense , Dropout , Flatten , Conv2D , MaxPooling2D
from keras.utils import np_utils
from keras.callbacks import EarlyStopping , ModelCheckpoint

#Scikit-learn
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , recall_score , precision_score , accuracy_score

#Pythonモジュール
import os
import glob
%matplotlib inline
```

```python
#画像データのパス指定
test_path = '/content/drive/MyDrive/Colab Notebooks/-codexa-画像処理入門/data/chest_xray/chest_xray/test/'
train_path = '/content/drive/MyDrive/Colab Notebooks/-codexa-画像処理入門/data/chest_xray/chest_xray/train/'
```

```python
# Testデータの読み込み
test_df = []
for foldername in os.listdir(test_path):
    imgs_path = test_path + foldername
    imgs = sorted(glob.glob(imgs_path + "/" + '*.jpeg'))
    
    for name in imgs:
        test_df.append((str(name), str(foldername)))

# データフレームへ格納
test_df = pd.DataFrame(test_df, columns=['Image', 'Label'], index=None)
test_df.head()
```

```python
#Trainデータの読み込み
train_df = []
for foldername in os.listdir(train_path):
    imgs_path = train_path + foldername
    imgs = sorted(glob.glob(imgs_path + '/' + '*.jpeg'))

    for name in imgs:
        train_df.append((str(name) , str(foldername)))

#データフレームへ格納
train_df = pd.DataFrame(train_df , columns=['Image' , 'Label'] ,index=None)
train_df.head()
```

```python
#メモをクリア
del X_train
del y_train
del X_test
del y_test

#npyファイルを読み込み
X_train = np.load(train_path + 'X_train.npy')
y_train = np.load(train_path + 'y_train.npy')
X_test = np.load(test_path + 'X_test.npy')
y_test = np.load(test_path + 'y_test.npy')

#サイズの確認
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Sequentualインスタンス生成
model = Sequential()

#入力層と畳み込み層
model.add(Conv2D(32 , kernel_size=(3,3),
                 activation='relu',
                 input_shape=(64,64,3)))

#プール層の追加
model.add(MaxPooling2D(pool_size=(2,2)))

#ドロップアウトの追加
model.add(Dropout(0.25))

#全結合層
model.add(Flatten())
model.add(Dense(128,activation='relu'))

#ドロップアウト
model.add(Dropout(0.33))

#出力層
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
#EarlyStoppingの設定
callbacks = EarlyStopping(monitor='val_loss' , patience = 5)
```

```python
#モデル訓練
model.fit(X_train , y_train , validation_data=(X_test , y_test) , callbacks=[callbacks] , epochs=100)
```

```python
#ベースラインモデル　テストデータで推測結果出力
base_y_pred = model.predict(X_test)
base_y_pred[0:5]
```

```python
#混合行列を算出
confusion_matrix(y_test ,base_y_pred, labels=[1,0])
```

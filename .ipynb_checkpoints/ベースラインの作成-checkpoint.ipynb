```python


#仮説：過学習により分析精度が上がらないのではないか

#dropoutを実装

#Sequentialモデルの生成
model_2 = Sequential()

#1層目 96*96
model_2.add(Conv2D(64,kernel_size=(3,3),
                  activation='relu',
                  padding='same',
                  input_shape = (96,96,3)))
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(Dropout(0.5))

#2層目 48*48
model_2.add(Conv2D(128 , kernel_size=(3,3), activation='relu' , padding='same'))
model_2.add(MaxPooling2D(pool_size= (2,2)))
model_2.add(Dropout(0.5))

#2層目 24*24
model_2.add(Conv2D(256 , kernel_size=(3,3) , activation='relu' , padding='same'))
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(Dropout(0.5))

#3層目　12*12
model_2.add(Conv2D(512 , kernel_size=(3,3) , activation='relu' , padding='same'))
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(Dropout(0.5))

#4層目　6*6
model_2.add(Conv2D(1024 , kernel_size=(3,3) , activation='relu' , padding='same'))
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(Dropout(0.5))

#全結合層1
model_2.add(Flatten())
model_2.add(Dense(1000,activation='relu'))

#出力層
model_2.add(Dense(10,activation='softmax'))

#モデルコンパイル　確率的勾配効果法
model_2.compile(loss=losses.categorical_crossentropy,
               optimizer='sgd',
               metrics=['accuracy'])
model_2.summary()

#モデルの訓練
model_2.fit(x=X_train,y=y_train ,batch_size=32,validation_data=(X_test,y_test),epochs=10000,callbacks=[callbacks])

```

```python
#データをfloat型へ変換
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

#正規化
X_train /= 255
X_test /= 255
```

```python
#kerasで訓練を行う用にデータを整形(バッチ、高さ、幅、チャンネル)
#size変更
X_train = X_train.reshape(X_train.shape[0] , 96 , 96 , 3)
X_test = X_test.reshape(X_test.shape[0] , 96 , 96 , 3)

print(X_train.shape)
print(X_test.shape)
```

```python
#バイナリーのクラス行列へ変換
y_train = np_utils.to_categorical(y_train , len(label_master))
y_test = np_utils.to_categorical(y_test , len(label_master))
```

```python
#テストデータの確認
#x番目のレコードのテストデータ画像ファイルを読み込み
def check_test(x):
    image_2 = X_test[x]
    image_2 = cv2.cvtColor(image_2 , cv2.COLOR_BGR2RGB)
    #読み込んだ画像データを表示
    plt.imshow(image_2)
    print('Labels-',np.argmax(y_test[x]))
    print('Image Shape-',image_2.shape)
    plt.title('Image-Label' + str(label_master.iloc[np.argmax(y_test[x])]))
    plt.show()


#trainデータの確認
#x番目のレコードのテストデータ画像ファイルを読み込み
def check_train(x):
    image_2 = X_train[x]
    image_2 = cv2.cvtColor(image_2 , cv2.COLOR_BGR2RGB)
    #読み込んだ画像データを表示
    plt.imshow(image_2)
    print('Labels-',np.argmax(y_train[x]))
    print('Image Shape-',image_2.shape)
    plt.title('Image-Label' + str(label_master.iloc[np.argmax(y_train[x])]))
    plt.show()
```

```python
#ベースモデルとなるモデルの作成 VGG16を参考(0パディング追加)
#Sequentialモデルの生成
model1 = Sequential()

#1層目 96*96
#入力層兼畳み込み層を追加
model1.add(Conv2D(64,kernel_size=(3,3),
                  activation='relu',
                  input_shape = (96,96,3)))

#プール層を追加
model1.add(MaxPooling2D(pool_size=(2,2)))

#2層目 48*48
model1.add(Conv2D(128 , kernel_size=(3,3), activation='relu'))
model1.add(MaxPooling2D(pool_size= (2,2)))


#全結合層1
model1.add(Flatten())
model1.add(Dense(1000,activation='relu'))

#出力層
model1.add(Dense(10,activation='softmax'))
```

```python
#モデルコンパイル　確率的勾配効果法
model1.compile(loss=losses.categorical_crossentropy,
               optimizer='sgd',
               metrics=['accuracy'])
model1.summary()
```

```python
#EarlyStoppingの設定
callbacks = EarlyStopping(monitor='val_loss' , patience=5)
```

```python
#モデルの訓練
model1.fit(X_train , y_train , validation_data=(X_test , y_test) , callbacks=[callbacks] , epochs=10000)
```
